```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris
```

```python
# Load the iris dataset
iris = load_iris()

# Convert to a DataFrame for easier manipulation
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df['target_names'] = iris_df['target'].apply(lambda x: iris.target_names[x])
```

```python
# Display the first few rows of the DataFrame
print(iris_df.head())
```

```python
# Display summary statistics
print(iris_df.describe())
```

Descriptive statistics and data distribution:

```python

# Count of each target class
print(iris_df['target_names'].value_counts())
```

```python
# Check the data types
print(iris_df.dtypes)
```

```python
#in order to create Heatmap to visualize the correlation between features
#I will need to ensure that only numerical columns are used for correlation 
numeric_df = iris_df.select_dtypes(include=['float64', 'int64'])
```

```python
# Heatmap to visualize the correlation between features 
plt.figure(figsize=(10, 6)) 
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

This heatmap above provides a clear visual representation of how different features in the dataset are related to each other, 
which can be useful for feature selection, data analysis, and building predictive models.

Understanding the Heatmap

•	Color Gradient: The heatmap uses a color gradient ranging from blue (negative correlation) to red (positive correlation).
•	Correlation Values: The values range from -1 to 1:
o	1: Perfect positive correlation (features increase together)
o	-1: Perfect negative correlation (one feature increases as the other decreases)
o	0: No correlation

Specific Correlations in Your Dataset

•	Sepal Length and Sepal Width: Correlation of -0.12 (weak negative correlation)
•	Sepal Length and Petal Length: Correlation of 0.87 (strong positive correlation)
•	Sepal Length and Petal Width: Correlation of 0.82 (strong positive correlation)
•	Sepal Width and Petal Length: Correlation of -0.43 (moderate negative correlation)
•	Sepal Width and Petal Width: Correlation of -0.37 (moderate negative correlation)
•	Petal Length and Petal Width: Correlation of 0.96 (very strong positive correlation)

Insights from the Heatmap

1.	Petal Length and Petal Width:
    
o	This pair has the highest positive correlation (0.96), indicating that as petal length increases, petal width also increases.

2.	Sepal Length with Petal Length and Petal Width:
o	Both pairs show strong positive correlations (0.87 and 0.82, respectively), suggesting that as sepal length increases, both petal length and petal width increase as well.

3.	Sepal Width:
    
o	Shows weaker negative correlations with other features, indicating less direct relationships.

Diagonal Elements
•	All diagonal elements are 1, indicating each feature is perfectly correlated with itself (which is expected).


```python
# Pairplot to visualize pairwise relationships in the dataset 
sns.pairplot(iris_df, hue='target_names', markers=["o", "s", "D"])
plt.show()
```

This pairplot above is useful to me for exploring the relationships between features and 
understanding how well each feature differentiates between the Iris species.

Overview of the Pairplot
•	Diagonal Plots: These are histograms (or sometimes kernel density estimates) showing the distribution of each feature (sepal length, sepal width, petal length, petal width) for the different Iris species.
•	Off-Diagonal Plots: These are scatter plots showing the relationships between each pair of features.
Color and Marker Legend
•	Setosa: Blue circles (o)
•	Versicolor: Orange squares (s)
•	Virginica: Green diamonds (D)


Insights from the Pairplot

1.	Sepal Length vs. Sepal Width:
o	Setosa (blue circles) are well-separated from Versicolor (orange squares) and Virginica (green diamonds).
o	Versicolor and Virginica show some overlap, indicating these features alone might not be sufficient to distinguish them.

2.	Petal Length vs. Petal Width:
o	Clear separation between all three species, especially for Setosa.
o	Versicolor and Virginica show a linear relationship with some overlap but are distinguishable.


3.	Other Features:
o	Petal length and width have stronger discriminatory power compared to sepal length and width, as Setosa is distinctly separate in these plots.

Additional Observations
•	Distributions: The histograms on the diagonal help me see how each feature is distributed for the different species.
•	Outliers and Patterns: The scatter plots chelp identify outliers and patterns within each pair of features.


```python
# Histogram to show distribution of each feature
iris_df.hist(bins=20, figsize=(10, 8))
plt.show()
```

On the histogram above.
Understanding such distribution patterns can help in better classifying and analyzing the Iris species based on their morphological features.


Blow is  an interesting insight from the histogram plot:

    Insight on Petal Length
    
•	The histogram for petal length reveals two distinct peaks (bimodal distribution), one around 1.5 cm and another around 4.5 cm.
•	This suggests that there are two dominant groups of petal lengths within the dataset.
•	These groups likely correspond to different Iris species, with the shorter petal lengths belonging to Iris setosa and the longer petal lengths likely belonging to Iris versicolor and Iris virginica.
•	This bimodal distribution indicates that petal length is a significant feature for distinguishing between the different Iris species.
